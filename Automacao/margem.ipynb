```python
import openpyxl

def ajustar_preco_ideal(caminho_arquivo, nome_planilha):
    try:
        # Carregar a planilha
        wb = openpyxl.load_workbook(caminho_arquivo)
        ws = wb[nome_planilha]
    except Exception as e:
        print(f"Erro ao carregar o arquivo ou planilha: {e}")
        return

    # Definir a linha inicial e a coluna de onde os dados começam
    linha_inicial = 2  # Linha 2, assumindo que a linha 1 é o cabeçalho
    coluna_preco_necessario = 19  # Coluna S
    coluna_preco_ideal = 3       # Coluna C
    coluna_dif = 21               # Coluna U

    linha_atual = linha_inicial

    # Loop para percorrer todas as linhas com dados
    while True:
        # Verificar se há dados na coluna PRECO NECESSARIO (coluna S)
        preco_necessario = ws.cell(row=linha_atual, column=coluna_preco_necessario).value
        if preco_necessario is None:
            break  # Sai do loop se não houver mais dados (caso PRECO NECESSARIO seja vazio)

        # Verificar o valor de DIF (coluna U)
        dif = ws.cell(row=linha_atual, column=coluna_dif).value
        if dif is None:
            dif = 0  # Se o DIF não estiver preenchido, assume-se 0
        
        # Exibir status da linha que está sendo processada
        print(f"Processando linha {linha_atual}...")

        # Ajustar o PRECO IDEAL até que o DIF seja 0
        while dif != 0:
            # Atualiza o PRECO IDEAL com o valor de PRECO NECESSARIO
            ws.cell(row=linha_atual, column=coluna_preco_ideal, value=preco_necessario)

            # Recalcular o valor de DIF após a mudança
            dif_recalculado = ws.cell(row=linha_atual, column=coluna_dif).value
            if dif_recalculado is None:
                dif_recalculado = 0  # Caso o valor de DIF seja None, considera como 0

            # Verificar se o valor de DIF realmente mudou
            if dif_recalculado == dif:
                # Se o DIF não mudar após a alteração, sai do loop
                print(f"DIF não foi alterado para linha {linha_atual}. Continuando para próxima linha.")
                break

            # Atualizar o valor do DIF para a próxima iteração
            dif = dif_recalculado

        # Quando o DIF for 0, exibe o status de conclusão da linha
        print(f"Linha {linha_atual} processada com sucesso.")

        # Passa para a próxima linha
        linha_atual += 1

    # Salvar as alterações feitas no arquivo
    wb.save(caminho_arquivo)
    print("Processo concluído.")

# Exemplo de uso
caminho_arquivo = 'entrada_produtos.xlsx'  # Substitua pelo caminho do seu arquivo Excel
nome_planilha = 'Planilha1'           # Substitua pelo nome correto da sua planilha
ajustar_preco_ideal(caminho_arquivo, nome_planilha)
```
